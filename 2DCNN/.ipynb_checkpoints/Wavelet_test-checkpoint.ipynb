# 2D CNN Aproach with scalogram

## Imports and setup

```python
import csv
import os

import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

from ssqueezepy import cwt, icwt
from ssqueezepy.visuals import plot, imshow
```

## Processing data

```python
df = pd.read_csv('../data/thm/sensors.csv')
df[["MB_temp","BMB_temp","B1_temp","B2_temp"]] = np.nan

df["MB_temp"] = df["temperature"][df["sensor_id"]==26]
df["BMB_temp"] = df["temperature"][df["sensor_id"]==27]
df["B1_temp"] = df["temperature"][df["sensor_id"]==28]
df["B2_temp"] = df["temperature"][df["sensor_id"]==29]

aggregation_functions = {'MB_temp': 'sum', 'BMB_temp': 'sum', 'B1_temp': 'sum',  'B2_temp': 'sum'}
df = df.groupby(df['created_on']).aggregate(aggregation_functions)
df.replace(0, np.nan, inplace=True)
df.fillna(method='pad', inplace=True)
```

## Preparing data for scalogram creation

```python
df["B1_temp_adj"] = df["B1_temp"]
df["B1_temp_adj"][df["B1_temp_adj"]>40] = np.nan
df.fillna(method='pad', inplace=True)

X=  df["B1_temp"].iloc[-100000:].values
X_adj=  df["B1_temp_adj"].iloc[-100000:].values
```

```python
X_test = X[2000:5000]

fig = plt.figure(figsize =(25, 8)) 
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax.plot(X_test)
ax.set_title("Original input data")
plt.show()

fig.savefig("original_input_data", facecolor='white', transparent=False)
```

```python
fig, axs = plt.subplots(4)
fig.set_figheight(20)
fig.set_figwidth(25)
fig.suptitle('Scalograms')

X_test = X[2000:5000]
Wx, scales = cwt(X_test, 'morlet')
Wx_imag = Wx.copy()
Wx_real = Wx.copy()
Wx_all = Wx.copy()

imag = np.where(Wx_imag.imag>0.1)
for i in range(0, len(imag[0])):
    Wx_imag[imag[0][i], imag[1][i]] = 0
    Wx_all[imag[0][i], imag[1][i]] = 0

real = np.where(Wx_real.real>0.1)
for i in range(0, len(real[0])):
    Wx_real[real[0][i], real[1][i]] = 0
    Wx_all[real[0][i], real[1][i]] = 0
    
imshow(Wx,ticks=0, borders=0, abs=1, ax=axs[0], show=0, title="Original wavelet")
imshow(Wx_imag,ticks=0, borders=0, abs=1, ax=axs[1], show=0, title="Imaginary adjusted wavelet")
imshow(Wx_real,ticks=0, borders=0, abs=1, ax=axs[2], show=0, title="Real Adjusted wavelet")
imshow(Wx_all,ticks=0, borders=0, abs=1, ax=axs[3], show=0, title="Real+Imag adjucted wavelet")

fig.savefig("wavelets", facecolor='white', transparent=False)
```

```python
fig, axs = plt.subplots(4)
fig.set_figheight(15)
fig.set_figwidth(25)
fig.suptitle('Chart and Scalogram')

X_rec = icwt(Wx, wavelet='morlet')
X_rec_imag = icwt(Wx_imag, wavelet='morlet')
X_rec_real = icwt(Wx_real, wavelet='morlet')
X_rec_all = icwt(Wx_all, wavelet='morlet')

axs[0].plot(X_rec)
axs[0].set_title("Original data")

axs[1].plot(X_rec_imag)
axs[1].set_title("Imaginary adjusted data reconstructed")

axs[2].plot(X_rec_real)
axs[2].set_title("Real adjusted data reconstructed")

axs[3].plot(X_rec_all)
axs[3].set_title("Adjusted data reconstructed")

plt.show()

fig.savefig("data_reconstructed_from_wavelets", facecolor='white', transparent=False)
```

```python
fig, axs = plt.subplots(2)
fig.set_figheight(10)
fig.set_figwidth(25)
fig.suptitle('Chart and Scalogram')
axs[0].plot(X_rec-X_rec_imag)
axs[0].plot(X_rec)
```
